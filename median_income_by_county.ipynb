```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = 'illinois_median_income_county.csv'
df = pd.read_csv(file)
df = df.iloc[:,0:3]
df['Median Income'] = df['Median Income'].str.replace(",","").astype(float)
counties = df['County'].unique()
```

```python
fig, ax = plt.subplots(10,1)
fig.set_figheight(30)
fig.set_figwidth(15)
fig.subplots_adjust(hspace = .5, wspace=.5)
num = 1

for county in counties:

    median_income = df.loc[(df['County']==county),'Median Income']
    year = df.loc[(df['County']==county),'Year']
    ax=plt.subplot(10,1,num)
    plt.grid()
#     ax.get_xaxis().set_visible(False)
#     ax.get_yaxis().set_visible(False)
#         ax.get_xaxis().set_ticks(years)
#         ax.get_yaxis().set_ticks([])

    ax.plot(year, median_income,color='b')
#         ax = fig.add_subplot(5,5,num)
#         ax.plot(year, race_married)
    ax.set_title(f"{county}")
    num = num + 1
```

```python
year = df.loc[(df['County']=='Cook'),'Year']


plt.plot(year,df.loc[(df['County']=='Cook'),'Median Income'],label='Cook')
plt.plot(year,df.loc[(df['County']=='Will'),'Median Income'],label='Will')
plt.plot(year,df.loc[(df['County']=='Lake'),'Median Income'],label='Lake')
plt.plot(year,df.loc[(df['County']=='DuPage'),'Median Income'],label='DuPage')
plt.plot(year,df.loc[(df['County']=='Pope'),'Median Income'],label='Pope')
plt.plot(year,df.loc[(df['County']=='Hardin'),'Median Income'],label='Hardin')
plt.plot(year,df.loc[(df['County']=='Kane'),'Median Income'],label='Kane')
plt.plot(year,df.loc[(df['County']=='Gallatin'),'Median Income'],label='Gallatin')
plt.plot(year,df.loc[(df['County']=='Calhoun'),'Median Income'],label='Calhoun')
plt.plot(year,df.loc[(df['County']=='Scott'),'Median Income'],label='Scott')
plt.title('Median Income by County')
plt.ylabel('Median Income ($)')
plt.xlabel('Year')
plt.grid()
plt.legend()
```
